```python
from functools import reduce
from itertools import chain, combinations

import numpy
import pandas
```

```python
d = "../dc.xlsx"
data = pandas.read_excel(d)
data
```

```python
data["_category_tag_split"] = data["category_tag"].str.split(" ")
data["_category_tag_split"] = data["_category_tag_split"].apply(
    func=lambda x: list(chain.from_iterable(combinations(x, n + 1) for n in range(len(x) - 1))),
)
data
```

```python
data.iloc[0, -1]
```

```python
list_of_all_non_unique = reduce(
    lambda x, y: x + y,
    data["_category_tag_split"].values.tolist(),
    [],
)
list_of_all_non_unique = [" ".join(list(x)) for x in list_of_all_non_unique]
list_of_all_unique = numpy.unique(list_of_all_non_unique).tolist()
list_of_all_unique = [x.split(" ") for x in list_of_all_unique]
list_of_all_unique
```

```python
def isin(list_of_tags, string_in_table):
    result = 1 if all([x in string_in_table for x in list_of_tags]) else 0
    return result

def count_in(list_of_lists_of_tags, list_of_strings_in_table):
    occurrences = [sum([isin(x, y) for y in list_of_strings_in_table]) for x in list_of_lists_of_tags]
    return occurrences
```

```python
occurrences = [numpy.isin(x, data["category_tag"].values.tolist()).sum() for x in list_of_all_unique]
occurrences
```

```python
summary = count_in(list_of_all_unique, data["category_tag"].values.tolist())
summary = pandas.DataFrame(
    data={
        "tags": list_of_all_unique,
        "no": summary,
    },
)
summary = summary.sort_values(by="no", ascending=False)
summary
```
